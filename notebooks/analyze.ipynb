```python
import pandas as pd
from tqdm.auto import tqdm
```

```python
df = pd.read_csv(
    "/Users/cthoyt/.data/indra/db/processed_statements-2022-03-31-embeddings.tsv.gz",
    sep="\t",
)
df.head()
```

```python
total = len(df.index)
total
```

```python
synonyms_idx = df.node.str.startswith("text:")
grounded_idx = ~synonyms_idx
```

```python
synonyms_path = "/Users/cthoyt/.data/indra/db/processed_statements-2022-03-31-synonyms.txt"
with open(synonyms_path, "w") as file:
    for s in tqdm(sorted(df.node[synonyms_idx]), unit_scale=True):
        print(s.removeprefix("text:").strip(), file=file)
```

```python
print(
    f"{grounded_idx.sum():,} ({grounded_idx.sum()/total:.2%}) are grounded, "
    f"{synonyms_idx.sum():,} ({synonyms_idx.sum()/total:.2%}) are not"
)
```
